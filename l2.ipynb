```python
def average(*numbers) : 
    sum = 0
    count = 0
    for i in numbers : 
        sum += i 
        count += 1 
    
    return sum / count
```

```python
average(3, 4, 6, 9)
```

3. Siljiydigan ortacha qiymat (Moving Average)
Tasavvur qiling, sizda songgi yetti kun davomida qayd etilgan murojaatlar soni boyicha malumotlar mavjud: [1, 2, 10, 3, 2, 4, 5]. Siz ushbu davr uchun ortacha murojaatlar darajasini aniqlamoqchisiz. Outliers  bu boshqa qiymatlardan keskin farq qiluvchi qiymatlar. Bunday outliers bilan kurashish usullaridan biri  ularni boshqa qiymatlar bilan almashtirishdir. Shunday usullardan biri  siljiydigan ortacha qiymat.

1-element uchun: siljiydigan ortacha mavjud emas, chunki hozircha faqat bitta qiymat bor.

2-element uchun: siljiydigan ortacha = (1 + 2) / 2 = 1.5

3-element uchun: siljiydigan ortacha = (2 + 10) / 2 = 6 va hokazo.
